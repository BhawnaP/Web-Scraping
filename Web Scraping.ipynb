```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
from pandas import Series, DataFrame
```

```python
url="https://www.ucop.edu/operating-budget//budgets-and-reports/legislative-reports/2017-18-legislative-session.html"
```

```python
result=requests.get(url)
c=result.content
```

```python
soup=BeautifulSoup(c)
```

```python
summary=soup.find('div',{'class':'span8 dotted-top', 'role':'main'})
tables=summary.find_all('table')
```

```python
data=[]
rows=tables[0].findAll('tr')
for tr in rows:
    cols=tr.findAll('td')
    
    for td in cols:
        text=td.find(text=True)
        print(text)
        data.append(text)
```

```python
data
```

```python
reports=[]
date=[]
index=0

for item in data:
    if item==None:
        #print(item)
        reports.append(data[index-1])
        date.append(data[index-2])
    index+=1
```

```python
date=Series(date)
reports=Series(reports)

legislative_df=pd.concat([date, reports], axis=1)
legislative_df.columns=['Date','Report']

legislative_df.drop([34],axis=0, inplace=True)
legislative_df
```

```python
legislative_df.to_excel('legislative_df.xlsx')
```
